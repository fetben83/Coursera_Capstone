## Data

To build our machine learning model we need data about apartments for sale in algiers city with their caracteristics. Those data was scipted from the most important advertissement platform in algeria called <a href="https://www.ouedkniss.com/">OuedKniss</a>.
The challenge was so high and we were able to extract informations from unstructured data sources and put them in a structured DB Table containing:
- Price,
- Total area,
- The floor,
- Number of rooms,
- The district.

By using the district collected from the OuedKniss web site, we tried to get geo coordinates  by using GeoAPIfy as a second phase.
The last phase of data collectction consost of using geo coordinates collected in the previous step as an input using Foursqure API to get number of restaurants around each position.

At the end of this phase of data collection, we had a structed table containing:
- 2268 rows representing number of advertissements,
- 8 features.
